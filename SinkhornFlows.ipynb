```python
# Imports
import torch
from sjko import *


# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(device)
```

```python
domain = torch.tensor([[-.5],
                       [0],
                       [1]], dtype=torch.get_default_dtype())
# domain = torch.eye(3)
# domain = grid[0:1:.1, 0:1:.1]
µ0 = torch.tensor([0.6, .1, .3])
# µ0 = uniform(domain)
# µ0 = diracs(domain, (0,-1))
V = lambda x: torch.tensor([.5, 1, 0.])
# V = lambda x: sqnorm(x)
# noncvx = lambda t: 1 - torch.cos(3*torch.pi*t) + t*t
# V = lambda x: noncvx(sqnorm(x))
F = PotentialEnergy(domain, V)
tau = 5e-3
N = 2000
time = tau*torch.arange(N)
lr = .1
descent_tol = 0
sinkhorn_tol = 1e-6
descent_maxiter = 40
sinkhorn_maxiter = 30
eps=.5
eps_list = [.1]
```

```python
fig = go.Figure(data=[potential_heatmap(V, domain)], layout=dict(height=700, width=700, scene=dict(aspectmode='cube')))
fig.show()
```

```python
flow = SJKO_flow(µ0, F, time, eps, descent_tol=descent_tol, descent_maxiter=descent_maxiter, lr=lr, sinkhorn_tol=sinkhorn_tol, sinkhorn_maxiter=sinkhorn_maxiter, verbose=True)
```

```python
fig = flow.go_anim_b(10, dt=10, fig_side_px=1000, axisvisible=False, B_kwargs=dict(line=dict(width=5, color='red')), traj_kwargs=dict(color='green', line=dict(width=10), marker=dict(size=20)))
fig.show()
```
